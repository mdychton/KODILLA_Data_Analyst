#USUWANIE WIERSZY - DROP + inplace=True

```python
df.drop('semestr1',axis=1)
```

```python
df
```

Po wykonaniu tej operacji wywołajmy jeszcze raz zmienną df w komórce poniżej, a zauważmy, że kolumna w dalszym ciągu tam jest i ma się dobrze!Dzieje się tak, ponieważ wykonując funkcję drop, nie ustawiliśmy wartości True dla parametru Inplace. Parametr ten jest swego rodzaju zaworem bezpieczeństwa, który pozwala nam wykonać jakąś operację bez rzeczywistej ingerencji we wprowadzoną wcześniej tabelę. Musimy w tym miejscu powiedzieć, że wiele funkcji Pandas działa właśnie w ten sposób, a parametr Inplace będzie przewijał się w trakcie tego szkolenia również przy innych modyfikacjach DataFrame.

Usuńmy zatem kolumny z tabeli przypisanej do zmiennej df raz na zawsze. Jednak tym razem na potrzeby przykładu razem z kolumną semester1 usuńmy kolumnę e3. W tym celu, jako pierwszy argument funkcji drop musimy wprowadzić listę kolumn do usunięcia.

```python
df
```

```python
df.drop(['e3','semestr1'],axis=1,inplace=True)
```

```python
df
```

```python
df.drop('Student B',inplace=True)
df
```
