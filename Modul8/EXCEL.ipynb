```python
writer = pd.ExcelWriter('add_chart.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='my_chart')
```

```python
workbook = writer.book
worksheet = writer.sheets['my_chart']
```

```python
chart = workbook.add_chart({'type':'line'})
```

```python
def grab_series(df, sheet_name, colname, startcol=0, startrow=0):

    col_index = df.columns.tolist().index(colname)
    col_letter = chr(ord('@')+(col_index+2+startcol))
    first_row = startrow + 2
    last_row = startrow + 1 + len(df)
    return f"='{sheet_name}'!{col_letter}{first_row}:{col_letter}{last_row}"
```

```python
chart.add_series({'values':grab_series(df,'my_chart','B')})
```

```python
chart.set_x_axis({
    'name': 'x^2',
    'name_font': {'size': 14, 'bold': True},
    'num_font':  {'italic': True },
})
```

```python
chart.set_legend({'none': True})
```

```python
worksheet.insert_chart('F2', chart)
```

```python
workbook.close()
```

```python
df.pivot_table(values='Sprzedaż',index='Przedstawiciel',columns='Region',aggfunc=np.sum) #TU PIVOT 
```

```python
df.pivot_table(values='Sprzedaż',index='Przedstawiciel',columns='Region',aggfunc=np.sum).fillna(0).round(2)
```

```python
df.pivot_table(values='Sprzedaż',index=['Region','Przedstawiciel'],aggfunc=np.sum).round(2).sort_values(by=['Region','Sprzedaż'],ascending=[True,False])
```

```python
df.pivot_table(values='Sprzedaż',index='Region',aggfunc=[len,np.max,np.min]).round(0) # Tym razem będziemy chcieli poznać liczbę, maksymalną oraz minimalną wartość transakcji w rozróżnieniu na poszczególne regiony.
```

# APPLY

```python
def commission_fee(x):
    if x <= 300:
        return 0
    elif x < 900: 
        return x * 0.03
    else:
        return x * 0.06
```

```python
df['commission_fee'] = df['Sprzedaż'].apply(lambda x: commission_fee(x))
```

```python
df
```

```python
df['Produkt_len'] = df['Produkt'].apply(len)
```

```python
df
```

```python
import math
```

```python
df['#_opakowan'] = df['Sztuki'].apply(lambda x: math.ceil(x/5))
```

```python
df
```

```python
def bonus(row):

    margin = (row['Sprzedaż'] - row['Koszty'])/row['Sprzedaż']

    if margin > 0.55:
        return 200
    else:
        return 0
    
```

```python
df['Bonus'] = df.apply(lambda row: bonus(row),axis =1)
```

```python
df.sample(10)
```
