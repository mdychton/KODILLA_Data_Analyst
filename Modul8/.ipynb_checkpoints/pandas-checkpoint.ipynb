#USUWANIE WIERSZY - DROP + inplace=True

```python
df.drop('semestr1',axis=1)
```

```python
df
```

Po wykonaniu tej operacji wywołajmy jeszcze raz zmienną df w komórce poniżej, a zauważmy, że kolumna w dalszym ciągu tam jest i ma się dobrze!Dzieje się tak, ponieważ wykonując funkcję drop, nie ustawiliśmy wartości True dla parametru Inplace. Parametr ten jest swego rodzaju zaworem bezpieczeństwa, który pozwala nam wykonać jakąś operację bez rzeczywistej ingerencji we wprowadzoną wcześniej tabelę. Musimy w tym miejscu powiedzieć, że wiele funkcji Pandas działa właśnie w ten sposób, a parametr Inplace będzie przewijał się w trakcie tego szkolenia również przy innych modyfikacjach DataFrame.

Usuńmy zatem kolumny z tabeli przypisanej do zmiennej df raz na zawsze. Jednak tym razem na potrzeby przykładu razem z kolumną semester1 usuńmy kolumnę e3. W tym celu, jako pierwszy argument funkcji drop musimy wprowadzić listę kolumn do usunięcia.

Jeśli wywołamy funkcję Groupby i wskażemy tylko kolumnę, według której chcemy wykonać grupowanie, otrzymamy z powrotem jedynie informację o miejscu w pamięci, gdzie możemy znaleźć obiekt Groupby. Ćwiczenie
Połącz ze sobą df1, df2, df3 wzdłuż osi poziomej używając funkcji concat. Postaraj się, aby połączona tabela nie rozszerzała się na dodatkowe wiersze w dół.

```python


# reset indexów
df1_reset = df1.reset_index(drop=True)
df2_reset = df2.reset_index(drop=True)
df3_reset = df3.reset_index(drop=True)

# concat poziomy
result = pd.concat([df1_reset, df2_reset, df3_reset], axis=1)

print(result)
```

-reset_index(drop=True) → ustawia index 0,1,2,3 dla wszystkich
-axis=1 → łączenie w poziomie (kolumny)
-dzięki temu pandas nie dodaje nowych wierszy

# MERGE

```python
def make_keys(n):
    return [f'key{n}' for n in range(n)]
left = df1.copy()
left['key'] = make_keys(4)
right = df3.copy()
right['key'] = make_keys(4)
```

```python
print(left)
```

```python
print(right)
```

```python
pd.merge(left,right,how="inner",on='key')  #innerJoin
```

```python
new_right=pd.DataFrame({'E':['E1','E2','E3','E4','E5','E6'],'key':make_keys(6)})
```

```python
new_right
```

```python
pd.merge(left,new_right,how='inner',on='key') #laczmy tabele left i tabele new_right - sposob inner , klucz key
```

```python
pd.merge(left,new_right,how='outer',on='key') #tutaj mamy outer - czyli bierzemy wszystko z obu tabel,jesli nie czegos w jednej to dostajemy NaN - w sensie nie ma tych samych kluczy ( elementow laczacych)
```

# JOIN

Funkcja join działa w zbliżony do merge sposób. Jako jedyna z trzech omawianych w tym rozdziale funkcji jest wykonywana bezpośrednio na obiekcie DataFrame, do którego przyłączana ma być inna tabela.

Trzeba jednak pamiętać, że przyłączana tabela będzie się łączyć zawsze po swoim indeksie. Jednakże kluczem z tabeli, do której będziemy przyłączać inną może być dowolna kolumna.

```python
other = pd.DataFrame({'E':['E1','E2','E3','E4','E5','E6']},index=['B1','B2','B3','B4','B5','B6'])
```

```python
print (other)
print (left)
```

Zauważmy, że left i other nie łączy wspólny indeks. Ich punktem wspólnym są z kolei kolumna ‘B’ w left oraz indeks w other. W takim wypadku możemy użyć funkcji join, aby je ze sobą połączyć.

```python
left.join(other,on='B') #left jest domyślny
```

```python
left.join(other,on='B',how='inner')
```
