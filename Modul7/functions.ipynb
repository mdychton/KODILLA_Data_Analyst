# Funkcje w Pythonie

Aby stworzyć funkcję, używamy słowa kluczowego `def`, a następnie:
- podajemy nazwę funkcji
- w nawiasach przekazujemy argumenty
- po dwukropku wpisujemy kod funkcji


```python
def add_numbers(a, b):
    return a + b
    
```

```python
# Wywoływanie funcji

Aby wywołać funcje, wpisujemy jej nazwe are przekazujemy paremtry w nawiasach okrągłych.
    
```

```python
result = add_numbers(2,8)
print (result)
```

##Możemy używać też komend magicznych


```python
%time add_numbers(1000,2000)
```
